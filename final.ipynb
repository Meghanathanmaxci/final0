```python
class Food_Order_App:
    def __init__(self):
        self.adminLogin   = False
        self.adminDetail  = {}
        self.userLogin    = False
        self.userDetail   = {}
        self.foods        = {}
        self.orderHistory = []
        
        
    def Register_Admin(self):
        self.admin_name = input("Enter the Name : ")
        a = int(input("Enter the mobile number :"))
        if len(str(a)) ==10:
            self.admin_number = a
        else:
            print("please enter valid number with 10 digits...")
            a = int(input("Enter the valid mobile number : "))
            if len(str(a)) == 10:
                self.admin_number = a
            
        self.admin_mailid   = input("Enter the mail_id :")
        self.admin_password = input('Create strong password')
        a = {}
        a["admin_name"]     = self.admin_name
        a["admin_mailid"]   = self.admin_mailid
        a["admin_number"]   = self.admin_number
        a["admin_password"] = self.admin_password
        
        
        self.adminDetail[self.admin_number] = a
        print("Registration completed sucessfully...")
        
        
    def Login_Admin(self):
        num = int(input("Enter your number :"))
        if num in self.adminDetail:
            b = input("Enter your password : ")
            if self.adminDetail[num]["admin_password"] == b:
                self.adminLogin = True
                print("login sucessfull...")
            else:
                print("Invalid password...")
        else:
            print("Register First or Enter valid Details to login...")
            
            
    def Add_New_item(self):
        self.food = []
        self.Food_Id = input("Create food_Id : ")
        c = input("Enter food name :")
        self.food.append(c)
        d = input("Enter the quantity of food :")
        self.food.append(d)
        e = int(input('Enter the price : '))
        self.food.append(e)
        self.foods[self.Food_Id] = self.food
        print("item is sucessfully added...")
        print(self.foods)
        
        
    def Discount(self):
        f = input("10% of discount available on every item you order, please order quicky : ")
        if "y" or "Y":
            print(self.foods.keys())
            for i in self.foods.keys():
                print(i)
            self.food[2] =self.food[2]-self.food[2]//10
            print("Discount applied order now")
            
            
            
    def Stock(self):
        print("available items are ",len(list(self.foods.keys())))
        k = 1
        for i in self.foods:
            print("available items are below and its details")
            print("S.No - ",k,"\n")
            print('Food_Id :',i)
            print("Food_Name :",self.foods[i][0])
            print("Food_Quantity : ",self.foods[i][1])
            print("Food_Price : ",self.foods[i][2])
            k+=1
            
            
            
    def Edit_item(self):
        g = input("Enter your Food_Id :")
        if g in self.foods:
            h = int(input("\n 1 To change Food_Id \n 2 To change Food_Name \n 3 To change Food_Quantity \n 4 To change Food_Price"))
            if h ==1:
                i = self.foods[g]
                self.foods.pop(g)
                j = input("Enter new Food_Id : ")
                self.foods[j] = i
                print("Food_Id is changed sucessfully...")
            if h ==2:
                self.foods[g][0]= input("Enter new Name for item : ")
                print("item name is changed sucessfully...")
            if h == 3:
                self.foods[g][1]= input("Enter the new Quantity for item:")
                print("item Quantity is changed sucessfully...")
            if h==4:
                self.foods[g][2]=int(input("Enter the new price for item :"))
                print("item price is changed sucessfully...")
        else:
            print("Invalid Food_Id...Sorry")
                
                
    def List_Of_Food_Items(self):
        print("Available items are : ",self.foods.values())
        
            
    def Remove_Items(self):
        k = input("Enter the Food_Id :")
        if k in self.foods:
            l = input("Do you wanna remove item which are not in stock..., if yes Enter Y or y")
            if l =="y" or "Y":
                self.foods.pop(k)
                print("list after removing item was updated...")
                if l == "n" or "N":
                    pass
            else:
                print("Please enter valid Food_Id...Thanks")
                
                
    def User_Register(self):
        print("Give us your details for Registration ")
        self.user_Name = input("Enter your name : ")
        m = int(input("Enter your mobile number :"))
        if len(str(m)) == 10:
             self.user_number = m
        else:
            print("Please enter 10 digits valid mobile number...")
        self.user_mailId   = input("Enter your mailId :")
        self.user_password = input("Create strong password :")
        self.user_place    = input("Enter your place : ") 
        n = {}
        n["user_Name"]     = self.user_Name
        n["user_number"]   = self.user_number
        n["user_mailId"]   = self.user_mailId
        n["user_password"] = self.user_password
        n["user_place"]    = self.user_place
        self.userDetail[self.user_number] = n
        print("Registration completed sucessfully...Thank you")
            
    def User_Login(self):
        num = int(input("Enter the mobile number : "))
        if num in self.userDetail:
            o = input("Enter your password : ")
            
            if self.userDetail[num]["user_password"]==o:
                self.userLogin =True
                print("login sucessfull...")
                p = int(input("\n 1 place new order \n 2 order history \n 3 update profile"))
            
                
                if p==1:
                    print("items available and its details ",list(self.foods.values()))
                    print("order by selecting the above option with food_Id as followed",self.foods)
                    q = input("Enter Food_Id :")
                    
                    if q in self.foods:
                        r = input("Confirm with y or Y to place order : ")
                        
                        if r =="y" or "Y":
                            s = self.foods.pop(q)
                            self.orderHistory.append(s)
                            print("order placed sucessfully...")
                    else:
                        print("choose vaild food_Id...")
                if p==2:
                    print("ordered details are saved here :",self.orderHistory)
                    
                if p==3:
                    u = int(input("Enter your mobile number : "))
                    
                    if u in self.userDetail:
                        v = list(self.userDetail.values())[0]
                        w = int(input("\n 1 To change name \n 2 To change mail_Id \n 3 To change number \n 4 To change password \n 5 To change place"))
                        
                        
                        if w==1:
                            v["user_name"] = input("Enter new name")
                            print("profile updated with new name sucessfully...",self.userDetail)
                        
                        if w==2:
                            v["user_mail_Id"] = input("Enter new mail_Id")
                            print("profile updated with new mail_Id sucessfully...",self.userDetail)
                        
                        if w==3:
                            v["user_number"] = input("Enter new number")
                            print("profile updated with new mobile number sucessfully...",self.userDetail)
                        
                        if w==4:
                            v["user_password"] = input("Enter new password")
                            print("profile updated with new password sucessfully...",self.userDetail)
                        
                        if w==5:
                            v["user_place"] = input("Enter new place")
                            print("profile updated with new place sucessfully...",self.userDetail)
                    
                    else:
                        print("choose vaild option...")
                                
       
        else:
            print("Invalid password")
    
    
    def Menu(self):
        x = True
        while x:
            y = int(input("\n 1 To Register_Admin \n 2 To Login_Admin \n 3 To Update_Item \n 4 Discount \n 5 To check stock \n 6 Edit Items \n 7 Available item \n 8 Remove Item \n 9 User Register \n 10 user Login"))
            
            if y ==1:
                self.Register_Admin()
            
            if y ==2:
                self.Login_Admin()
            
            if y ==3:
                self.Add_New_item()
            
            if y ==4:
                self.Discount()
            
            if y ==5:
                self.Stock()
            
            if y ==6:
                self.Edit_item()
            
            if y==7:
                self.List_Of_Food_Items()
            
            if y==8:
                self.Remove_Items()
            
            if y==9:
                self.User_Register()
            
            if y==10:
                self.User_Login()
        
        else:
            print("Enter valid option...")
```
